```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
df=pd.read_csv(r"Sales Data.csv")
df.head()
```

```python
df.drop({"Unnamed: 0"},axis=1,inplace=True)
```

```python
## Descriptive Analysis.
df_pv=pd.pivot_table(df,index="Product",columns="City",values=["Sales"],aggfunc="sum")
df_pv
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
```

```python
df['Month Name'] = df["Order Date"].dt.month_name()
```

```python
## Calculate the Revenue Generated
df["Revenue"]=df["Quantity Ordered"] * df["Price Each"].sum()
df
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x=df["Month Name"],color="Red")
plt.show()
```

```python
fig1=px.treemap(df,path=["City","Product","Month Name"],values="Price Each")
fig1.update_layout(title="Treemap of Total Prices by City, Product, and Month.",
                  paper_bgcolor="lightBlue",
                  plot_bgcolor="White")
fig1.show()
```

```python
fig1.write_html("scatter_plot.html")
```
